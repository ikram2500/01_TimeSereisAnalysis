## Time Resampling

```python
df
```

```python
# resample data on annual basis
df.resample('A').min()
```

```python
 df.resample('A').max()
```

```python
# year end frequency
df.resample(rule='A').max()['Open'].plot()
```

```python
df.resample(rule='QS').max()['High'].plot()
```

```python
# Business end frequency
df.resample(rule='BA').max()['High'].plot()
```

```python
df.resample(rule='BA').max()
```

```python
df.resample(rule='BQS').max()
```

```python
df.resample(rule='BQS').max().plot()
```

```python
df['Open'].resample(rule='BA').mean().plot(kind='bar')
```

```python
df['Open'].resample(rule='M').max().plot(kind='bar', figsize=(15,6))
```

# Rolling

# Aggregate Function

```python
df['Open 30 days rolling']= df['Open'].rolling(30).mean()
```
